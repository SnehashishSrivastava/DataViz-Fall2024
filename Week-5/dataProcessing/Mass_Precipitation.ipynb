Data source:<br>
https://www.mass.gov/info-details/precipitation-data

```python
df = pd.read_excel('Precipitation Database 10.19.xls')
```

## Example 6.4 North Dartmouth Precipitation 

```python
dfDartmouth = df[df["CITY"]=="North Dartmouth"].copy()
dfDartmouth = dfDartmouth[dfDartmouth.columns[1:14]].copy()
dfDartmouth.set_index("YEAR",drop=True,inplace=True)
dfDartmouth.head()
```

```python
dfDartmouth.to_csv("Dartmouth.csv");
```

## Exercise All Cities

```python
df_allCities = df.dropna()[df.columns[:-3]]
df_allCities.set_index("YEAR",inplace=True,drop=True)
df_allCities['STATION']=df_allCities.apply(lambda x: x['STATION']+"_"+ x['CITY'],axis=1)
```

```python
df_allCities[df_allCities.columns[:-1]].head()
```

```python
df_allCities[df_allCities.columns[:-1]].to_csv('Mass_Precipitation.csv')
```
